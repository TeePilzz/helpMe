```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the web page to scrape
url = "https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%8E%D0%B7_%D0%A1%D0%B5%D1%80%D0%BE%D0%BA%D1%80%D1%8B%D0%BB%D1%8B%D1%85"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the paragraphs (assuming the text is in <p> tags)
    paragraphs = soup.find_all('p')

    # Extract the text from the paragraphs
    text_data = [paragraph.get_text(strip=True) for paragraph in paragraphs]

    # Remove empty strings from the list
    text_data = [text for text in text_data if text]

    # Create a DataFrame to store the text data
    df = pd.DataFrame(text_data, columns=['Text'])

    # Save the DataFrame to an Excel file
    df.to_excel('director_script_text.xlsx', index=False)

    print("Text data has been successfully scraped and saved to 'director_script_text.xlsx'")
else:
    print(f"Failed to retrieve the web page. Status code: {response.status_code}")
```
